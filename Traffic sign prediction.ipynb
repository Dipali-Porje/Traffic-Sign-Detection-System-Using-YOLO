```python
import cv2

# Start video capture from the webcam (0 is default for the first connected webcam)
cap = cv2.VideoCapture(0)

while True:
    # Read a frame from the webcam
    ret, frame = cap.read()

    if not ret:
        print("Failed to grab frame")
        break

    # Run prediction on the frame
    results = model(frame)  # Perform detection on the current frame

    # Extract predictions
    frame = results.render()[0]  # Draw bounding boxes and labels on the frame

    # Display the frame with the traffic sign detection
    cv2.imshow('Traffic Sign Detection', frame)

    # Exit if the 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close any OpenCV windows
cap.release()
cv2.destroyAllWindows()
```

```python
import zipfile
import os

# Define the folder you want to zip
folder_name = 'runs'  # Replace with your folder name
zip_filename = 'runs.zip'  # Desired output filename

# Create a zip file
with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for root, dirs, files in os.walk(folder_name):
        for file in files:
            zipf.write(os.path.join(root, file), os.path.relpath(os.path.join(root, file), folder_name))

print(f"Zipped {folder_name} to {zip_filename}")
```
