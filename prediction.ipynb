Predict Traffic Sign 

```python
#import libraries
from ultralytics import YOLO
import cv2
```

```python
#define model
model = YOLO('best_traffic.pt')
```

```python
#detect images
def detect(image):
    result = model(image)
    return result
```

```python
#main function for continuously capturing images from camera or webcam
def main():
    cap = cv2.VideoCapture(0)
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        result = detect(frame)
        frame = result[0].plot()
        cv2.imshow("frame", frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()
```

```python
#run main function
if __name__ == "__main__":
    main()
```

```python
import cv2
import pyttsx3
from ultralytics import YOLO
import time

# Initialize TTS engine
engine = pyttsx3.init()
engine.setProperty('rate', 150)  # Adjust speech rate

# Load YOLO model
model = YOLO('C:/Users/porje/OneDrive/Desktop/internship jb infotech/yolo/traffic sign detection/best_traffic.pt')

# Initialize webcam
cap = cv2.VideoCapture(0)  # 0 for default webcam

# Store last detection timestamps to avoid constant repetition
last_detection_time = {}
announcement_delay = 50  # Announce the same sign again only after 3 seconds

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO prediction
    results = model(frame)

    detected_labels = []

    for result in results:
        boxes = result.boxes
        for box in boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = float(box.conf[0])
            cls = int(box.cls[0])

            # Get class name
            label = model.names[cls]

            # Draw bounding box
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, f'{label} {conf:.2f}', (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

            # Check last detection time
            current_time = time.time()
            if label not in last_detection_time or (current_time - last_detection_time[label]) > announcement_delay:
                detected_labels.append(label)
                last_detection_time[label] = current_time  # Update detection time

    # Speak detected traffic signs
    for label in detected_labels:
        print(f"Detected: {label}")
        engine.say(f"Detected traffic sign is {label}")
        engine.runAndWait()

    # Show the frame
    cv2.imshow('Real-Time Traffic Sign Detection', frame)

    # Press 'q' to quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
